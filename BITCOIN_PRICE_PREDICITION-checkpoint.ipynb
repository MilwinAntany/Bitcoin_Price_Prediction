# 1.Loading data as a CSV file:

# 2.Data Cleaning :

 # 2.1. Changing the column name:

```python
list(df.columns)
```

The column name is in wrong format like:
    i) The first letter is capital.
    ii) The bracketed values.
So, To "Avoid Case-Sensitivity Issues" we need to change it 
 


```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(r"\(.*\)", "", regex=True)
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.rstrip('_')
```

```python
list(df.columns)
```

Now, the column is ready for use 

  #  2.2.Finding if any duplicates exists :

```python
df.duplicated()
```

 There is no Duplicates exists in the dataset

   #  2.3.Finding the datatype:

```python
df.info()
```

 Here,  a column "date" is in incorrect format. so we have change into a proper datetime format using Pandas to_datetime() function.


```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
```

Then, there is more digits in the column Closing Price (USD), 24h Open (USD), 24h High (USD), 24h Low (USD). 
So, now round off them into two digits

```python
df['closing_price'] = df['closing_price'].round(2)
```

```python
df.head(10)
```

```python
df.info()
```

Now, Eveything is good for the next step...

# 3. Building a Machine Learning model for prediction :

Here, Linear Regression is used, because it provides a simple yet effective way to model the relationship between historical price data and future predictions.


```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


X = df[['24h_open', '24h_high', '24h_low']]
Y = df['closing_price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

model = LinearRegression()
model.fit(X_train, Y_train)
```

So, We have sucessfully build a Machine Learing model 


# 4.Checking the accuracy of The Model :


```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(Y_test, prediction)
mse = mean_squared_error(Y_test, prediction)
r2 = r2_score(Y_test, prediction)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Square Error: {mse}")
print(f"Root Mean Squared Error(R²): {r2}")

```

# We have 0.99 (R²), which is Good for The Model
